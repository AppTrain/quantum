```python
def amplitude_probabilities(a,b):
    import numpy as np 
    import warnings
    warnings.simplefilter("ignore", np.ComplexWarning)
    # if a != np.conj(a):
    a = sqrt((a*np.conj(a)).real)
    # if b != np.conj(b):
    b = sqrt((b*np.conj(b)).real)
    mag = sqrt(a**2 + b**2)
    return a/mag,b/mag, (a/mag)**2,(b/mag)**2
```
